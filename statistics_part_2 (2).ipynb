
### 1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on a sample of data. It involves testing a claim or hypothesis about a population parameter (such as a mean or proportion) using sample data. The goal is to determine whether there is enough evidence to reject or fail to reject the null hypothesis.

### 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- **Null hypothesis (H₀):** It is the hypothesis that there is no effect or no difference in the population. It represents a default or starting assumption.
- **Alternative hypothesis (H₁ or Ha):** It is the hypothesis that contradicts the null hypothesis. It suggests that there is an effect or difference in the population.

**Difference:** The null hypothesis represents the "status quo" or no effect, while the alternative hypothesis represents the claim we are testing.

### 3. What is the significance level in hypothesis testing, and why is it important?
- The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true (i.e., making a Type 1 error). Common values for α are 0.05, 0.01, and 0.10.
- It is important because it sets the threshold for how much evidence we need to reject the null hypothesis. A smaller α requires stronger evidence to reject H₀.

### 4. What does a P-value represent in hypothesis testing?
- The **P-value** is the probability of observing the sample data (or more extreme data) under the assumption that the null hypothesis is true. It measures the strength of the evidence against the null hypothesis.

### 5. How do you interpret the P-value in hypothesis testing?
- If the **P-value is less than or equal to the significance level (α)**, we reject the null hypothesis, suggesting strong evidence for the alternative hypothesis.
- If the **P-value is greater than α**, we fail to reject the null hypothesis, suggesting insufficient evidence for the alternative hypothesis.

### 6. What are Type 1 and Type 2 errors in hypothesis testing?
- **Type 1 error (False Positive):** Rejecting the null hypothesis when it is actually true.
- **Type 2 error (False Negative):** Failing to reject the null hypothesis when it is actually false.

### 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- **One-tailed test:** Tests for the possibility of an effect in one direction (either greater than or less than).
- **Two-tailed test:** Tests for the possibility of an effect in either direction (greater than or less than).

### 8. What is the Z-test, and when is it used in hypothesis testing?
The **Z-test** is used when the population variance is known or the sample size is large enough to approximate the population distribution. It is used to test hypotheses about population means or proportions.

### 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- The **Z-score** is calculated as:
  \[
  Z = \frac{X - \mu}{\sigma / \sqrt{n}}
  \]
  where \(X\) is the sample mean, \(\mu\) is the population mean, \(\sigma\) is the population standard deviation, and \(n\) is the sample size.
- The Z-score represents how many standard deviations the sample mean is from the population mean. It helps determine whether the result is statistically significant.

### 10. What is the T-distribution, and when should it be used instead of the normal distribution?
The **T-distribution** is used instead of the normal distribution when the sample size is small (typically less than 30) or when the population variance is unknown. It is more spread out and has heavier tails than the normal distribution.

### 11. What is the difference between a Z-test and a T-test?
- A **Z-test** is used when the population variance is known, or the sample size is large.
- A **T-test** is used when the population variance is unknown and the sample size is small.

### 12. What is the T-test, and how is it used in hypothesis testing?
The **T-test** is used to compare the means of two groups or to compare a sample mean to a population mean when the sample size is small and the population variance is unknown.

### 13. What is the relationship between Z-test and T-test in hypothesis testing?
Both tests are used for comparing means, but the **Z-test** is used when the population variance is known or the sample size is large, while the **T-test** is used when the population variance is unknown and the sample size is small.

### 14. What is a confidence interval, and how is it used to interpret statistical results?
A **confidence interval** provides a range of values within which the true population parameter is likely to fall, with a certain level of confidence (e.g., 95% confidence). It is used to estimate the precision of a sample statistic.

### 15. What is the margin of error, and how does it affect the confidence interval?
The **margin of error** is the range above and below the sample statistic in the confidence interval. A larger margin of error means the interval is wider and less precise.

### 16. How is Bayes' Theorem used in statistics, and what is its significance?
**Bayes' Theorem** provides a way to update the probability of a hypothesis based on new evidence. It is significant in statistical inference, especially in fields like machine learning and decision theory.

### 17. What is the Chi-square distribution, and when is it used?
The **Chi-square distribution** is used for categorical data to assess how well observed frequencies match expected frequencies. It is often used in tests like the Chi-square goodness-of-fit test and the Chi-square test for independence.

### 18. What is the Chi-square goodness of fit test, and how is it applied?
The **Chi-square goodness-of-fit test** is used to determine if a sample data matches a specific distribution. It compares observed frequencies to expected frequencies to see if there's a significant difference.

### 19. What is the F-distribution, and when is it used in hypothesis testing?
The **F-distribution** is used primarily in **ANOVA** (Analysis of Variance) and regression analysis to test the equality of variances across different groups or to compare model fits.

### 20. What is an ANOVA test, and what are its assumptions?
**ANOVA** (Analysis of Variance) is used to compare means across multiple groups. Assumptions include:
- Independence of observations
- Normally distributed data
- Equal variances (homogeneity of variance)

### 21. What are the different types of ANOVA tests?
- **One-way ANOVA:** Compares means of three or more independent groups based on one factor.
- **Two-way ANOVA:** Compares means across multiple groups considering two factors (with or without interaction).

### 22. What is the F-test, and how does it relate to hypothesis testing?
The **F-test** is used to compare two variances or to compare the fits of different models. It is the basis for ANOVA and tests whether the variability between group means is greater than the variability within groups.


```python
# @title Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results
import numpy as np
from scipy import stats

# Sample data
sample_data = np.array([100, 102, 98, 105, 107, 101, 100])  # Sample data (observations)
population_mean = 100  # Known population mean
population_std_dev = 5  # Known population standard deviation
sample_size = len(sample_data)  # Size of the sample

# Calculate the sample mean
sample_mean = np.mean(sample_data)

# Perform the Z-test (one-sample)
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print the Z-score and p-value
print(f'Sample Mean: {sample_mean}')
print(f'Z-score: {z_score}')
print(f'P-value: {p_value}')

# Interpretation based on a significance level (alpha)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
# @title Q 5 .Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def type_error_visualization(population_mean, population_std_dev, sample_size, alpha=0.05, effect_size=1):
    """
    Visualizes Type I and Type II errors during hypothesis testing.

    Args:
    - population_mean (float): The population mean under the null hypothesis.
    - population_std_dev (float): The population standard deviation.
    - sample_size (int): The sample size.
    - alpha (float): The significance level (Type I error rate).
    - effect_size (float): The true difference between population mean and the true mean under the alternative hypothesis.

    Returns:
    - None: Displays a plot with Type I and Type II errors.
    """

    # 1. Calculate critical value (rejection region boundary) for Type I error
    critical_value = stats.norm.ppf(1 - alpha)

    # 2. Calculate the distribution of the sample mean under the null hypothesis
    null_mean = population_mean
    null_std = population_std_dev / np.sqrt(sample_size)

    # 3. Calculate the distribution of the sample mean under the alternative hypothesis (true mean)
    true_mean = population_mean + effect_size
    alt_std = population_std_dev / np.sqrt(sample_size)

    # 4. Type I Error: Rejecting the null hypothesis when it is true (Area under the rejection region)
    rejection_area_start = critical_value
    rejection_area_end = np.inf

    # 5. Type II Error: Failing to reject the null hypothesis when it is false (Area to the left of the critical value)
    beta = stats.norm.cdf(critical_value, loc=true_mean, scale=alt_std)

    # Create a range of x values for plotting
    x = np.linspace(null_mean - 4 * null_std, null_mean + 4 * null_std, 1000)

    # Plot the normal distribution under the null hypothesis (H0)
    plt.plot(x, stats.norm.pdf(x, null_mean, null_std), label="Null Hypothesis (H0)", color="blue")

    # Plot the normal distribution under the alternative hypothesis (H1)
    plt.plot(x, stats.norm.pdf(x, true_mean, alt_std), label="Alternative Hypothesis (H1)", color="orange")

    # Shade the rejection region (Type I error region)
    plt.fill_between(x, 0, stats.norm.pdf(x, null_mean, null_std), where=(x > critical_value), color="red", alpha=0.3, label="Rejection Region (Type I Error)")

    # Shade the Type II error area (Beta area)
    plt.fill_between(x, 0, stats.norm.pdf(x, true_mean, alt_std), where=(x < critical_value), color="green", alpha=0.3, label="Type II Error Region (Beta Area)")

    # Annotate the critical value
    plt.axvline(x=critical_value, color="black", linestyle="--", label=f"Critical Value (Z = {critical_value:.2f})")

    # Labels and title
    plt.title("Type I and Type II Error Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

    # Output Type I and Type II errors
    print(f"Type I Error (alpha): {alpha}")
    print(f"Type II Error (beta): {beta:.4f}")
    print(f"Power of the test (1 - Type II Error): {1 - beta:.4f}")

# Example usage
population_mean = 100
population_std_dev = 5
sample_size = 30
alpha = 0.05  # Type I error rate (significance level)
effect_size = 1  # True difference between population mean and alternative hypothesis mean

# Visualize Type I and Type II errors
type_error_visualization(population_mean, population_std_dev, sample_size, alpha, effect_size)
```

```python
# @title Q6 .Write a Python program to perform an independent T-test and interpret the results
import numpy as np
from scipy import stats

# Example sample data for two groups
group_1 = np.array([23, 21, 18, 24, 25, 19, 20, 22, 21, 24])  # Sample data for Group 1
group_2 = np.array([30, 32, 31, 33, 29, 35, 30, 28, 32, 31])  # Sample data for Group 2

# Perform the independent t-test
t_statistic, p_value = stats.ttest_ind(group_1, group_2)

# Print the results
print(f"Group 1 Data: {group_1}")
print(f"Group 2 Data: {group_2}")
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Hypothesis testing interpretation
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The means of the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: The means of the two groups are not significantly different.")
```

```python
# @title Q 9Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy import stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Args:
    - sample_data (array): The sample data for which to calculate the CI.
    - confidence_level (float): The confidence level (default is 95%).

    Returns:
    - (CI_lower, CI_upper): The lower and upper bounds of the confidence interval.
    """
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Use sample std (ddof=1)
    sample_size = len(sample_data)

    # Calculate the standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate the margin of error using the t-distribution (for small samples)
    confidence_interval = stats.t.interval(confidence_level, df=sample_size-1, loc=sample_mean, scale=standard_error)

    return confidence_interval

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Sample data with mean 50 and std 10

# Calculate the 95% confidence interval
CI_lower, CI_upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: ({CI_lower:.2f}, {CI_upper:.2f})")
```

```python
# @title Q 10 Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy import stats

def margin_of_error(sample_data, confidence_level=0.95):
    """
    Calculate the margin of error for a sample mean given a confidence level.
    Args:
    - sample_data (array): Sample data.
    - confidence_level (float): Desired confidence level (default 0.95).

    Returns:
    - margin of error.
    """
    # Sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    # Z-score for the desired confidence level
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    # Calculate the margin of error
    margin_error = z_score * (sample_std / np.sqrt(sample_size))

    return margin_error

# Example
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Sample data with mean 50 and std 10

MoE = margin_of_error(sample_data, confidence_level=0.95)
print(f"Margin of Error: {MoE:.2f}")
```

```python
# @title Q.11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior, likelihood, evidence):
    """
    Apply Bayes' Theorem to calculate the posterior probability.

    Args:
    - prior (float): Prior probability P(H).
    - likelihood (float): Likelihood P(E|H).
    - evidence (float): Probability of evidence P(E).

    Returns:
    - posterior (float): Posterior probability P(H|E).
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example
prior = 0.6  # Prior probability of hypothesis H
likelihood = 0.8  # Likelihood of evidence given H
evidence = 0.75  # Probability of evidence

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior Probability: {posterior:.3f}")
```

```python
# @title Q.12 Perform a Chi-square test for independence between two categorical variables in Python
import pandas as pd
from scipy.stats import chi2_contingency

# Example data: Let's assume we have two categorical variables: 'Gender' and 'Preference'
data = {
    'Male': [30, 10],  # 30 males prefer A, 10 males prefer B
    'Female': [20, 20],  # 20 females prefer A, 20 females prefer B
}

# Convert the data into a DataFrame
df = pd.DataFrame(data, index=['Prefer A', 'Prefer B'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(df)

# Output the results
print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:\n{expected}")

# Interpretation of the results
if p < 0.05:
    print("Reject the null hypothesis: There is a significant relationship between the variables.")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship between the variables.")
```

```python
# @title Q.13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies based on observed data for a Chi-square test.
    observed: 2D array or list of lists representing the observed frequencies.
    Returns: 2D array of expected frequencies.
    """
    # Convert the observed data to a NumPy array for easier manipulation
    observed = np.array(observed)

    # Calculate the total number of observations (N)
    N = np.sum(observed)

    # Calculate the row sums (R) and column sums (C)
    row_sums = np.sum(observed, axis=1)  # Sum along rows
    col_sums = np.sum(observed, axis=0)  # Sum along columns

    # Initialize an empty array to store the expected frequencies
    expected = np.zeros_like(observed, dtype=float)

    # Calculate the expected frequencies for each cell
    for i in range(observed.shape[0]):  # Iterate over rows
        for j in range(observed.shape[1]):  # Iterate over columns
            expected[i, j] = (row_sums[i] * col_sums[j]) / N

    return expected

# Example observed data
observed_data = [
    [30, 10],  # Male: 30 prefer A, 10 prefer B
    [20, 20]   # Female: 20 prefer A, 20 prefer B
]

# Calculate expected frequencies
expected_data = calculate_expected_frequencies(observed_data)

# Display the expected frequencies
print("Observed Frequencies:")
print(np.array(observed_data))

print("\nExpected Frequencies:")
print(expected_data)
```

```python
# @title  Q.14 Perform a goodness-of-fit test using Python to compare the observed data to an expected distributionD
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected):
    """
    Perform a Chi-square goodness-of-fit test to compare observed data to expected distribution.
    observed: 1D array or list of observed frequencies.
    expected: 1D array or list of expected frequencies.
    Returns: Chi-square statistic and p-value.
    """
    # Perform the Chi-square goodness-of-fit test using scipy
    chi2_stat, p_value = chisquare(observed, expected)
    return chi2_stat, p_value

# Example: Observed data and expected data (uniform distribution)
observed_data = [50, 60, 55, 35]  # Example observed frequencies
expected_data = [50, 50, 50, 50]  # Expected frequencies (uniform distribution)

# Perform the test
chi2_stat, p_value = goodness_of_fit_test(observed_data, expected_data)

# Output the results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpretation of the results
if p_value < 0.05:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")

```

```python
# @title Q.15 Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Function to simulate and visualize the Chi-square distribution
def plot_chi_square_distribution(df_values, sample_size=10000):
    """
    Simulate and visualize the Chi-square distribution for different degrees of freedom.

    df_values: List or array of degrees of freedom to plot.
    sample_size: The number of random samples to generate for simulation.
    """
    plt.figure(figsize=(10, 6))

    # Plot the chi-square distribution for each degrees of freedom (df)
    x = np.linspace(0, 20, 500)  # Values to plot on the x-axis

    for df in df_values:
        # Generate random samples from the Chi-square distribution
        data = np.random.chisquare(df, sample_size)

        # Plot the theoretical Chi-square PDF for the given df
        plt.plot(x, chi2.pdf(x, df), label=f'df={df} (Theoretical)', color='black', linestyle='dashed')

        # Plot the histogram of the simulated data
        plt.hist(data, bins=50, density=True, alpha=0.6, label=f'df={df} (Simulated)')

    # Customize the plot
    plt.title("Chi-square Distribution with Different Degrees of Freedom")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Degrees of freedom to simulate
df_values = [2, 5, 10, 20]

# Call the function to plot the Chi-square distribution
plot_chi_square_distribution(df_values)
```

```python
# @title Q.16 Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
import scipy.stats as stats

# Generate two random samples
np.random.seed(0)
sample_1 = np.random.normal(loc=0, scale=1, size=100)  # Sample 1 (mean 0, std deviation 1)
sample_2 = np.random.normal(loc=0, scale=2, size=100)  # Sample 2 (mean 0, std deviation 2)

# Calculate sample variances
var_1 = np.var(sample_1, ddof=1)  # ddof=1 to get the sample variance (unbiased)
var_2 = np.var(sample_2, ddof=1)

# Calculate the F-statistic
F_statistic = var_1 / var_2

# Degrees of freedom
df_1 = len(sample_1) - 1
df_2 = len(sample_2) - 1

# Calculate the p-value from the F-distribution
p_value = 1 - stats.f.cdf(F_statistic, df_1, df_2)  # Upper-tail test

print(f"F-statistic: {F_statistic}")
print(f"p-value: {p_value}")
```

```python
# @title Q.19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    """
    Checks the assumptions for ANOVA: normality, independence, and equal variances.

    Parameters:
    groups (tuple of arrays): Group data to test assumptions.

    Returns:
    None: Outputs assumption checks.
    """

    # Check normality assumption using Shapiro-Wilk test
    print("Normality Assumption:")
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: Shapiro-Wilk Test p-value = {p_value}")
        if p_value < 0.05:
            print(f"Group {i} is not normally distributed (p < 0.05).")
        else:
            print(f"Group {i} is normally distributed (p >= 0.05).")

    print("\nVariance Assumption:")
    # Check equal variances using Levene's test (more robust to non-normality)
    stat, p_value = stats.levene(*groups)
    print(f"Levene's test p-value = {p_value}")
    if p_value < 0.05:
        print("The variances are significantly different (p < 0.05).")
    else:
        print("The variances are equal (p >= 0.05).")

    # Visualize the data distribution (boxplot for normality)
    print("\nBoxplot for Visualizing Data Distribution:")
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=groups)
    plt.title('Boxplot of Groups')
    plt.xlabel('Groups')
    plt.ylabel('Values')
    plt.show()

    # Note: Independence can't be directly tested with statistical tests
    print("\nIndependence Assumption:")
    print("Independence is assumed based on study design (e.g., random sampling).")


# Example usage with three sample groups
group_1 = np.random.normal(10, 2, 50)  # Sample group 1
group_2 = np.random.normal(12, 2, 50)  # Sample group 2
group_3 = np.random.normal(14, 2, 50)  # Sample group 3

# Check assumptions for the groups
check_anova_assumptions(group_1, group_2, group_3)
```

```python
# @title Q.24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

# 1. Simulate sample data (e.g., sample of size 30)
np.random.seed(42)

# Sample data generated from a normal distribution (Mean = 10, Std = 4)
sample_data = np.random.normal(10, 4, 30)

# 2. Define the population variance under the null hypothesis
sigma_squared_0 = 16  # Example: H0: population variance = 16 (standard deviation 4)

# 3. Calculate the sample variance (s^2)
sample_variance = np.var(sample_data, ddof=1)  # ddof=1 for sample variance

# 4. Calculate the test statistic (Chi-square statistic)
n = len(sample_data)  # sample size
chi_square_statistic = (n - 1) * sample_variance / sigma_squared_0

# 5. Perform the chi-square test (two-tailed test)
alpha = 0.05  # significance level
degrees_of_freedom = n - 1

# Calculate the critical values for the Chi-square distribution
chi_square_critical_lower = stats.chi2.ppf(alpha/2, degrees_of_freedom)  # Lower critical value (left tail)
chi_square_critical_upper = stats.chi2.ppf(1 - alpha/2, degrees_of_freedom)  # Upper critical value (right tail)

# 6. Compare the test statistic with the critical values
print(f"Chi-square statistic: {chi_square_statistic}")
print(f"Critical values: Lower = {chi_square_critical_lower}, Upper = {chi_square_critical_upper}")

# 7. Interpret the results
if chi_square_statistic < chi_square_critical_lower or chi_square_statistic > chi_square_critical_upper:
    print("Reject the null hypothesis: The population variance is different from 16.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from the population variance of 16.")

```

```python
# @title Q.25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

# Sample data for two groups
# Group 1: 80 successes out of 200 trials
x1 = 80  # successes
n1 = 200  # total trials

# Group 2: 70 successes out of 180 trials
x2 = 70  # successes
n2 = 180  # total trials

# 1. Calculate the sample proportions
p1 = x1 / n1
p2 = x2 / n2

# 2. Calculate the pooled proportion
p = (x1 + x2) / (n1 + n2)

# 3. Calculate the Z-statistic
z = (p1 - p2) / np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

# 4. Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# 5. Output the results
print(f"Sample Proportion Group 1: {p1}")
print(f"Sample Proportion Group 2: {p2}")
print(f"Pooled Proportion: {p}")
print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

# 6. Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the proportions.")
```

```python
# @title Q.27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Simulate some observed data (e.g., frequencies of categories A, B, C)
np.random.seed(42)

# Simulate 1000 observations with probabilities for categories A, B, C
observed_data = np.random.choice(['A', 'B', 'C'], size=1000, p=[0.5, 0.3, 0.2])

# 2. Observed frequencies for each category
observed_freq = [np.sum(observed_data == 'A'), np.sum(observed_data == 'B'), np.sum(observed_data == 'C')]

# 3. Define the expected frequencies (assume the expected proportions are 50%, 30%, and 20%)
total_observations = len(observed_data)
expected_freq = [total_observations * 0.5, total_observations * 0.3, total_observations * 0.2]

# 4. Perform the Chi-square goodness of fit test
chi2_statistic, p_value = stats.chisquare(observed_freq, expected_freq)

# 5. Output the results of the Chi-square test
print(f"Observed frequencies: {observed_freq}")
print(f"Expected frequencies: {expected_freq}")
print(f"Chi-square statistic: {chi2_statistic}")
print(f"P-value: {p_value}")

# 6. Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")

# 7. Visualization: Bar plot of observed vs expected frequencies
categories = ['A', 'B', 'C']
width = 0.3  # bar width

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(np.arange(len(categories)) - width/2, observed_freq, width=width, label='Observed', color='blue')
ax.bar(np.arange(len(categories)) + width/2, expected_freq, width=width, label='Expected', color='orange')

# Adding labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Frequencies')
ax.set_title('Observed vs Expected Frequencies')
ax.set_xticks(np.arange(len(categories)))
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```
